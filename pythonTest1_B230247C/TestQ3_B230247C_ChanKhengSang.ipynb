```python
import pandas as pd
import matplotlib.pyplot as plt

# prompt the system to read the csv file called coffee.csv
cf = pd.read_csv('coffee.csv')

# print out the title
print(cf.head())

# count the total transaction quantity of each product type
# calling the dataset columns such as 'product_category', 'product_type', and 'transaction_qty'
# Group by 'product_category' and 'product_type', then sum the 'transaction_qty' values
total_orders = cf.groupby(['product_category', 'product_type'])['transaction_qty'].sum().reset_index()

# Storing the results in a DataFrame
# It have already stored it in 'total_orders' DataFrame
print(total_orders)

# showing the data
# Pivot the DataFrame for better plotting
pivot_table = total_orders.pivot(index='product_type', columns='product_category', values='transaction_qty').fillna(0)

# Plotting
pivot_table.plot(kind='bar', figsize=(12, 8))
plt.title('Total Transaction Quantity of Each Product Type by Product Category')
plt.xlabel('Product Type')
plt.ylabel('Total Transaction Quantity')
plt.xticks(rotation=90)
plt.legend(title='Product Category')
plt.tight_layout()

# Save and show the plot
plt.savefig('total_transaction_quantity_plot.png')
plt.show()
```
